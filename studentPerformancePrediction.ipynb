```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#load the data
data=pd.read_csv('student-mat.csv', sep=';')
data.head()
```

```python
#feature selection
data=data[["G1","G2","G3","studytime","failures","absences"]]
```

```python
#print(data)
```

```python
#reorder data to have the x and y
x=np.array(data.drop(["G3"],1))
y=np.array(data["G3"])
#print(x)
```

```python
#split the data
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#fn to calculate grade
def getGrade(s):
    if s<=35:
        print(f"Score: {s} , Grade is F")
    elif s <=45:
        print(f"Score: {s} , Grade is D")
    elif s <=65:
        print(f"Score: {s} , Grade is C")
    elif s <=85:
        print(f"Score: {s} , Grade is B")
    else:
        print(f"Score: {s} , Grade is A")
```

```python
#prediction
result=model.predict(x_test)
```
